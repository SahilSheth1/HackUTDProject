```python
import pandas as pd
```

```python
filePath = ("HackUTD-2023-HomeBuyerInfo.csv")

df = pd.read_csv(filePath)
df.head()
```

```python
result = df[df["CreditScore"] > 640]

df['Good Credit'] = df['CreditScore'] > 640
df.head()
```

```python
df["LTV"] = (df["AppraisedValue"] - df["DownPayment"]) / df["AppraisedValue"]
df.head()
```

```python
pmiNeeded = df["LTV"] > 0.8
df["pmiNeeded"] = df["LTV"] > 0.8

df.loc[pmiNeeded, "AppraisedValue"] *= 1.01
df["MonthlyMortgagePayment"] = df["MonthlyMortgagePayment"].round(2)
df.head()
```

```python
df.head()
#   Calculate DTI (Debt to Income ratio) 
#   Monthly debt = CCPayment + CarPayment + StudentLoan + Mortgage payment


df["Monthly Debt"] = (df["CarPayment"] + df["MonthlyMortgagePayment"] + df["CreditCardPayment"])
df.head()


df['DTI'] = (df["Monthly Debt"] / df['GrossMonthlyIncome']).round(2)
df.head()
```

```python
df['FEDTI'] = (df['MonthlyMortgagePayment'] / df['GrossMonthlyIncome'])
df['FEDTI'] = df["FEDTI"].round(2)
df.head()
```

```python
approval = (df["LTV"] < 0.95) & (df["DTI"] < 0.43) & (df["FEDTI"] < 0.29)
df["Approved"] = approval
df.head(30)
```
