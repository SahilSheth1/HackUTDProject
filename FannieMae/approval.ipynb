```python
import pandas as pd
```

```python
filePath = ("HackUTD-2023-HomeBuyerInfo.csv")

df = pd.read_csv(filePath)
df.head()
```

```python
result = df[df["CreditScore"] > 640]

df['Good Credit'] = df['CreditScore'] > 640
df.head()
```

```python
df["Loan To Value"] = (df["AppraisedValue"] - df["DownPayment"]) / df["AppraisedValue"]
df.head()
```

```python



pmiNeeded = df["Loan To Value"] > 0.8
df["pmiNeeded"] = df["Loan To Value"] > 0.8

df.loc[pmiNeeded, "MonthlyMortgagePayment"] *= 1.01
df["MonthlyMortgagePayment"] = df["MonthlyMortgagePayment"].round(2)
df.head()
```

```python
df['FEDTI'] = (df['MonthlyMortgagePayment'] / df['GrossMonthlyIncome']) * 100
df.head()
```
